```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
```

```python
x = np.random.rand(10,10)#0~1
y = -1+x*2
print(y)#-1~1
plt.figure(figsize = (8, 6))
sns.heatmap(y, cmap = plt.cm.RdYlBu_r,
            vmin = -0.25, annot = True, vmax = 0.6)

plt.title('Correlation Heatmap')
```

```python
x1 = np.random.rand(1000,3)
y1 = -1+x1*2
indice = np.random.choice([0,1,2], size=1000)

y1_label = pd.DataFrame(y1,indice).reset_index()


grid = sns.PairGrid(data = y1_label, size = 3, diag_sharey=False,
                    hue = 'index',
                    vars = [x for x in list(y1_label.columns) if x != 'index'])


# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.distplot)

# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.suptitle('Ext Source and Age Features Pairs Plot', size = 32, y = 1.05)
plt.show()
```
